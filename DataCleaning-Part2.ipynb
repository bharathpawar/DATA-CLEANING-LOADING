# Filter with more then one condition using OR

```python
df = pd.read_csv('employees.csv', parse_dates = ["Start Date", "Last Login Time"])

df["Senior Management"] = df["Senior Management"].astype("bool")
```

```python
mask1 = df["Senior Management"]

mask2 = df["Start Date"] < "1990-01-01"
```

```python
df[mask1 | mask2]
```

```python
# It does not matter about start date ... If Any one condition is satisfied .. We will get the result
```

```python
df.columns
```

```python
df["First Name"] == "Robert"

maskf = df["First Name"] == "Robert"
```

```python
df["Team"] == "Client Services"

maskt = df["Team"] == "Client Services"
```

```python
df["Start Date"] > "2016-06-01"

maskd = df["Start Date"] > "2016-06-01"
```

```python
df[maskf & maskt | maskd]
```

```python
df[(maskf & maskt) | maskd]
```
